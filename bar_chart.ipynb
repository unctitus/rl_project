```python
import matplotlib.pyplot as plt

# Data
agent1_data = {'Win': 24, 'Draw': 0, 'Lost': 976}
agent2_data = {'Win': 32, 'Draw': 0, 'Lost': 968}

# Extracting values from dictionaries
agent1_values = list(agent1_data.values())
agent2_values = list(agent2_data.values())

# Labels for the bars
labels = list(agent1_data.keys())

# Bar positions
bar_width = 0.35  # Width of each bar
index = list(range(len(labels)))

# Plotting the bar chart
fig, ax = plt.subplots()
bar1 = ax.bar(index, agent1_values, bar_width, label='DQN Agent 1')
bar2 = ax.bar([i + bar_width for i in index], agent2_values, bar_width, label='DQN Agent 2')

# Adding labels, title, and legend
ax.set_xlabel('Outcome')
ax.set_ylabel('Episodes')
ax.set_title('Results Comparison of DQN Agents against MCTS')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(labels)
ax.legend()

# Show percentages on the bars
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.2f}%'.format(height / sum(agent1_values) * 100),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bar1)
autolabel(bar2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = {
    'Minimax Depth': list(range(1, 15)),
    'Agent 1 Outcome': [
        'Win', 'Draw', 'Draw', 'Win', 'Win', 'Win', 'Win', 'Win',
        'Win', 'Win', 'Win', 'Win', 'Draw', 'Win'
    ],
    'Agent 2 Outcome': [
        'Win', 'Win', 'Win', 'Win', 'Draw', 'Win', 'Win', 'Win',
        'Win', 'Win', 'Win', 'Win', 'Draw', 'Win'
    ]
}

# Creating the table
fig, ax = plt.subplots()
ax.axis('off')  # Hides the axis

table_data = []
for key, values in data.items():
    table_data.append([key] + values)

# Creating the table
table = ax.table(cellText=table_data, loc='center', cellLoc='center', colLabels=table_data[0], colColours=['lightgray'] * len(table_data[0]), cellColours=[['w']*len(table_data[0])] + [['lightblue']*len(table_data[0])] * (len(table_data) - 1))

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.5, 1.5)  # Adjust the table size if needed

plt.title('Minimax Depth and DQN Agent Outcomes')

plt.show()
```
